```python
import pandas as pd
import numpy as np

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
transaction_data = pd.read_csv('./Result.csv')
product = pd.read_csv('./product_Result.csv')
```

```python
df = transaction_data.copy()
df.shape
```

```python
product.head()
```

```python
df_1 = df.merge(product[['PRODUCT_ID','DEPARTMENT','COMMODITY_DESC','SUB_COMMODITY_DESC']])
```

# 購物籃分析(SUB_COMMODITY_DESC)

> 預先標籤有無購買產品  
> 要用哪個欄位做購物籃分析(COMMODITY_DESC/SUB_COMMODITY_DESC)?

```python
basket_plus = df_1.groupby(['BASKET_ID','SUB_COMMODITY_DESC'])['QUANTITY']\
               .sum().unstack().reset_index().fillna(0).set_index('BASKET_ID')
```

```python
print(basket_plus.shape)
basket_plus.head()
```

```python
# 一筆BASKET_ID中標籤有無購買產品
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    else:
        return 1
```

```python
basker_encode_plus = basket_plus.applymap(encode_units)
basker_encode_plus
```

```python
# 一筆BASKET_ID中不只購買一種產品
```

```python
basket_filter_plus = basker_encode_plus[(basker_encode_plus > 0).sum(axis=1) >= 2]
basket_filter_plus
```

### 應用 Apriori 算法  
https://medium.com/towards-artificial-intelligence/association-discovery-the-apriori-algorithm-28c1e71e0f04

### min_support? 0.03

```python
frequent_itemsets = apriori(basket_filter_plus, min_support=0.03, use_colnames=True)\
    .sort_values('support', ascending=False).reset_index(drop=True)
# 計算商品購買次數
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
# df_1[df_1.COMMODITY_DESC=='FLUID MILK PRODUCTS'].shape
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)\
    .sort_values('lift', ascending=False).reset_index(drop=True)
rules
```

```python
rules[ (rules['lift'] >= 6) & (rules['confidence'] >= 0.5) ]
```
