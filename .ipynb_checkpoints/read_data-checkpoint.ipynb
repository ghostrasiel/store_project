```python
dt = product_df.COMMODITY_DESC.unique()
print(sorted(dt))
```

```python
product_df.COMMODITY_DESC.unique()
unique_commodity = product_df.COMMODITY_DESC.nunique()
print('商品大類:',unique_commodity,'項')
```
